### A practical and intuitive guide to Machine Learning models

This project provides a structured and comparable overview
of common supervised machine learning models.

Each model is presented using the same pipeline and layout,
with the goal of understanding **what the model does,
why each step is needed,
and how to adapt it to real-world data**.



| Model                  | Learning Type | Task           | Output type                |
|------------------------|---------------|----------------|----------------------------|
| Linear Regression      | Supervised    | Regression     | Continuous value           |
| KNN Regression         | Supervised    | Regression     | Continuous value           |
| Random Forest Regression | Supervised | Regression     | Continuous value           |
| Logistic Regression    | Supervised    | Classification | Class + probability        |
| KNN Classification     | Supervised    | Classification | Class + probability        |
| Random Forest Classification | Supervised | Classification | Class + probability        |


### Regression
- The target variable is **continuous**
- The model predicts a numerical value
- Examples:
  - house price
  - temperature
  - income

### Classification
- The target variable is **categorical**
- The model predicts a class label (and often a probability)
- Examples:
  - spam vs not spam
  - disease vs healthy
  - fraud vs not fraud


### Standard machine learning pipeline

All models in this project follow the same high-level workflow:

1. Dataset loading
2. Train-test split
3. Feature scaling
4. Model initialization
5. Model training
6. Predictions
7. Model evaluation
8. Model persistence


### Project philosophy

The goal of ML-Methods is not just to run models,
but to understand them.

If you can:
- explain what the model does
- explain why each step exists
- adapt the pipeline to your own data

then you truly understand the model.
